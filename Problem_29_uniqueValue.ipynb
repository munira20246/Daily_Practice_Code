```python
# 29. Write a Python program to get unique values from a list.
uni = [1, 3, 5, 7, 9, 11, 5, 6, 2, 3, 7, 10]
unique = []

for i in uni:
    if i not in unique:
        unique.append(i)

print(unique)
```

```python
def  unique(x):
    unique = []

    for i in x:
      if i not in unique:
        unique.append(i)  
    print(sorted(unique))     

li = [1, 3, 5, 7, 9, 11, 5, 6, 2, 3, 7, 10]        
unique(li)  
```
