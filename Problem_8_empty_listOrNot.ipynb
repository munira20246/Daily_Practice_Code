```python
#8. Write a Python program to check a list is empty or not.
def Empty(a):
  not_empty = False
  
  for i in a:
    if i == i:
      print(f"This list {a} is not a empty list")
      not_empty = True
      break  

  if not not_empty:  
      print(f"This list {a} is a empty list")

k = []
Empty(k)
```
