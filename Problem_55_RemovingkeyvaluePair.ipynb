```python
# 55. Write a Python program to remove key values pairs from a list of dictionaries.
d = [{'key1': 'value1', 'key2': 'value2'}, 
                 {'key1': 'value3', 'key2': 'value4'}]
e = []

for i in d:
    f = {}      
    for j, v in i.items():
       if j == 'key2':
           f[j] = v
           break
    e.append(f) 
  
print(e)       
 
```
