```python
"""5. Write a Python program to count the number of strings where the string length is 2 or more
and the first and last character are same from a given list of strings."""

stri = ['abc', 'bb', 'cti', 'ohiyo', '2112', 'tsc', 'bcs', 'aira', '124', '123231', 'onio']
counterr = 0

for i in stri:
  if len(i) >= 2:
    if i[0] == i[-1]:
      counterr += 1
print(counterr)
```

```python
def stringCounter(x):
    counter = 0
    
    for i in x:
        if len(i) >= 2:
            if i[0] == i[-1]:
                counter += 1
    return(counter)

s = ['bb', 'cti', 'ohiyo', '2112', 'tsc', 'bcs', 'aira', '124', '123231']    
stringCounter(s)        
```
