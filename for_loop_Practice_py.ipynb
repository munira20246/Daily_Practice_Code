```python
a = int(input("Enter 1st nm----"))  # My special practice(vognangsher jog[duti shonkha])
b = int(input("Enter 2nd nm----"))
x = int(input("Enter 3rd nm----"))
y = int(input("Enter 4th nm----"))

def vogn(a, b, x, y):
  by = b * y
  ay = a * y
  bx = b * x
  return by, ay, bx

by, ay, bx = vogn(a, b, x, y)
subtr_aybx = ay - bx
sum_aybx = ay + bx
f_subtr = f"{subtr_aybx}/{by}"
f_sum = f"{sum_aybx}/{by}"
print(f_subtr)
print(f_sum)

```
