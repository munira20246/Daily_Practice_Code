```python
# 65. Write a Python program to move all zero 
# digits to end of a given list of numbers.

n = [3, 4, 0, 0, 0, 6, 2, 0, 6, 7, 6, 0, 0, 0, 
 9, 10, 7, 4, 4, 5, 3, 0, 0, 2, 9, 7, 1]
z = []
for i in n:
    if i == 0:
        continue
    z.append(i)
for j in n:
    if j != 0:
        continue 
    z.append(j)   
print(z)    
    
      
        
```
