```python
# 12. Write a Python program to print a specified list after removing the 0th, 4th and 5th elements.
fruits = ['apple', 'banana', 'pine-apple', 'water-melon', 'strawberry', 'mango', 'orange', 'malta']
fru_its = []
for i in fruits:
  if i == fruits[0] or i == fruits[4] or i == fruits[5]:
    continue
  fru_its.append(i)

print(fru_its)
```
