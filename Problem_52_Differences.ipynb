```python
# 52. Write a Python program to compute the difference between two lists.
co1 = ["red", "orange", "green", "blue", "white"]
co2 = ["black", "yellow", "green","blue"]
di1 = list(set(co1) - set(co2))
di2 = list(set(co2) - set(co1))
print("Color1-Color2",di1)
print("Color2-Color1",di2)
```

```python
# Expected Output:
# Color1-Color2: ['white', 'orange', 'red']
# Color2-Color1: ['black', 'yellow']
```
