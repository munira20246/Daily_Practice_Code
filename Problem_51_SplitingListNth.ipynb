```python
# 51. Write a Python program to split a list every Nth element.
s_l = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 
            'i', 'j', 'k', 'l', 'm', 'n']
f = []
step = int(input("Enter nth number: "))
for i in range(step):
    g = s_l[i::step]
    f.append(g)
print(f)   
```

```python

# Expected Output: [['a', 'd', 'g', 'j', 'm'], ['b', 'e', 'h', 'k', 'n'], ['c', 'f', 'i', 'l']
```
