```python
# 40. Write a Python program to split a list based on first character of word.
word_list = ['be', 'have', 'do', 'say', 'get', 'make', 
             'go', 'know', 'take', 'see', 'come','think',
             'look', 'want', 'give', 'use', 'find', 'tell', 
             'ask', 'work', 'seem', 'feel', 'leave', 'call']

result = {}

for item in word_list:
    first_char = item[0]
    if first_char not in result:
        result[first_char] = []
    result[first_char].append(item)

for letter in sorted(result):
    print(f"{letter.upper()}: {result[letter]}")
```
