```python
'''26. Write a python program to check whether two lists are circularly identical.'''
a = [2, 30, 5, 36, 49, 64]
b = [49, 64, 2, 30, 5, 36]

match_found = False

for i in range(len(a)):
  a = a[1:] + a[:1]
  if a == b:
    print(f"Two lists {a} and {b} are circularly identical.")
    match_found = True
    break

if not match_found:
  print(f"Two lists {a} and {b} are not circularly identical.")
```

```python
def circularlyIdentical(x, y):
    match_found = False

    for i in range(len(x)):
      x = x[1:] + x[:1]
      if x == y:
        print(f"Two lists {x} and {y} are circularly identical.")
        match_found = True
        break

    if not match_found:
       print(f"Two lists {x} and {y} are not circularly identical.")

a = [2, 30, 5, 36, 49, 64]
b = [30, 5, 36, 49, 64, 2]

circularlyIdentical(a, b)
```
