```python
# 7. Write a Python program to remove duplicates from a list

li = [2, 5, 9, 3, 1, 8, 6, 4, 7, 2, 2, 7, 5, 3, 9, 8, 6, 4, 3, 3, 5, 5, 6]

se = set(li)
li = list(se)
print(li)
```

```python
unique = []

for i in li:
  if i not in unique:
    unique.append(i)  
print(sorted(unique))    
```

```python
def  unique(x):
    unique = []

    for i in x:
      if i not in unique:
        unique.append(i)  
    print(sorted(unique))     

li = [2, 5, 9, 3, 1, 8, 6, 4, 7, 2, 2, 7, 5, 3, 9, 8, 6, 4, 3, 3, 5, 5, 6]        
unique(li)   
```
