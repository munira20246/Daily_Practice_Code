```python
# 42. Write a Python program to find missing and additional values in two lists.

a = [8, 3, 0, 1, 6, 'a', 'b', 'c']
b = [3, 1, 'b', 9, 10, 5, 4, 2]

print("The are the missing value: ")
for i in a:
    if i not in b:
        print(i, end=' ')
print("\nThere are the additional value: ")        
for j in b:
    if j not in a:
        print(j, end=' ')

```
