```python
''' 17. Write a Python program to generate and print a list except for the first 5 elements,
where the values are square of numbers between 1 and 30 (both included). '''
sqre_num = []

for i in range(1, 31):
  k = i ** 2
  sqre_num.append(k)
print(sqre_num[5:], end=' ')
```
