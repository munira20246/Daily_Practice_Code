```python
''' 6. Write a Python program to get a list,
sorted in increasing order by the last element in each tuple from a given list of non-empty tuples.'''
st = [(3,6), (2, 5), (1, 2), (4, 4), (2, 3), (2, 1), (4, 0)]


new_st = []
for i in range(10000):
  for j in st:
    if j[-1] == i:
      new_st.append(j)
      break

print("Sample_List :", st)
print()
print("List Made   :", new_st)
```

```python
def getting_list(ts):
  new_st = []
  for i in range(100000):
    for j in ts:
      if j[-1] == i:
        new_st.append(j)
        break
  return new_st

st = [(3,6), (2, 5), (1, 2), (4, 4), (2, 3), (2, 1), (4, 7)]
new_st = getting_list(st)
print("Sample_list  :", st)
print()
print("List_Made    :", new_st)
```

```python
def listSorting(y):
      l1 = []

      for item in y:
        for i in range(len(l1)):
           if item[1] < l1[i][1]:
              l1.insert(i, item)
              break
        else:
           l1.append(item) 
      return l1    
      
 
samp_lst = [(4, 8), (3, 6), (5, 10), (2, 1), (1, 4)]
list1 = listSorting(samp_lst)
print("Original List :", samp_lst)
print()
print("Sorted List   :", list1)
```
