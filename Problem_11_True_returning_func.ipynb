```python
#11. Write a Python function that takes two lists and returns True if they have at least one common member.

def if_common(a, b):
  c = set(a)
  d = set(b)
  dc = c & d
  if dc:
    return True
  else:
    return False

x = [5, 6, 2, 3, 7]
y = [12, 10, 6, 9, 8]
if_common(x, y)
```
