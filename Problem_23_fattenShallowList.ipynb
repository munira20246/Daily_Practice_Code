```python
# 23. Write a Python program to flatten a shallow list.

def flatten(a):
    li = []
    for sublist in a:
        if isinstance(sublist, list):
            li.extend(flatten(sublist))

        else:
            li.append(sublist)    

    return li
fl = [[2, 4, 3], [1, 5, 6], [9], [7, 9, 0]]
flatten(fl)        
```
