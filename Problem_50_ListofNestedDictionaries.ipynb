```python
# 50. Write a Python program to sort a list of nested dictionaries.

my_list = [{'key': {'subkey': 1}}, {'key': {'subkey': 10}}, {'key': {'subkey': 5}}]

print("Original List: ")
print(my_list)

my_list.sort(key=lambda e: e['key']['subkey'], reverse=True)
print("Sorted List: ")
print(my_list) 
```

```python
# If there are different keys in a dictionary-

my_list = [
    {'munira': {'subject': 1}},
    {'sopna': {'subject': 10}},
    {'jahan': {'subject': 5}}
]
my_list.sort(key=lambda e: list(e.values())[0]['subject'], reverse=True)
print('Sorted List:')
print(my_list)
```
