```python
"""38. Write a Python program to change the position of every n-th value with the
(n+1)th in a list."""
n = [0, 1, 2, 3, 4, 5]

for i in range(0, len(n) - 1, 2): 
        n[i], n[i + 1] = n[i + 1], n[i]
      
print(n)        
```
