```python
# 59. Write a Python program to check 
# whether the n-th element exists in a given list.
x = [1, 2, 3, 4, 5, 6, 8]
n = int(input("Enter nth number: "))
for i in range(len(x)):
    if n == i:
        print(f"Yes, {n}th element is in the list which is --{x[n]}")
    if n < 0 and n >= -(len(x)):
        xlen = len(x) + n
        print(f"Yes, ({n})th element is in the list which is -- {x[xlen]}")
        break
    
    if n >= len(x):
        print(f"{n}th element is not in the list")
        break

    if n < -(len(x)):
        print(f"{n}th element is not in the list")
        break
```
