```python
# 45. Write a Python program to convert a pair of values into a sorted unique array.

L = [(1, 2), (3, 4), (1, 2), (5, 6), (7, 8), (1, 2), (3, 4), 
     (3, 4), (7, 8), (9, 10)]
k = []
for i in L:
    for j in i:
        k.append(j)
k = set(k)
print(list(k))
```
