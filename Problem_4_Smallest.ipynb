```python
#4. Write a Python program to get the smallest number from a list.

lowest = [20, 5, 9, 10, 8, 6, 900, 4, 7]
mi = float('inf')
for i in lowest:
  if i < mi:
    mi = i
print(mi)

```

```python
def Minimum(xz):
  infinity_num = float('inf')
  for i in xz:
    if i < infinity_num:
      infinity_num = i
  print(f"The Smallest Number from the list {xz} is---{infinity_num}")

p = [20, 5, 9, 10, 8, 6, 900, 7]
Minimum(p)
```
