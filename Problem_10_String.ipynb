```python
# 10. Write a Python program to find the list of words that are longer than n from a given list of words.
lis = ['The', 'quick', 'brown', 'foxjumps', 'overthe', 'lazydog']
n = int(input("Enter nth number---"))

for i in lis:
  if len(i) > n:
    print(i)
```

```python
def findwords(x):
    n = int(input("Enter value of n: "))
    for i in x:
        if len(i) > n:
            print(i)

lis2 = ['Bangladesh', 'Indonesia', 'Malaysia', 'india', 'peru', 'Pakistan', 'kazakhistan']
findwords(lis2) 
```
