```python
# 30. Write a Python program to get the frequency of the elements in a list.

my_list = [1, 3, 5, 3, 7, 1, 3, 5, 9, 1]
frequency = {}

for item in my_list:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

print("Frequency of elements:", frequency)

```

```python
def gettingFreq(a):
    frequency = {}

    for item in a:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1
    print("Frequency of elements:", frequency)

my_list = [1, 3, 5, 3, 7, 1, 3, 5, 9, 1]
gettingFreq(my_list)             
```
