```python
""" 31. Write a Python program to count the number of elements in a list within a
specified range."""

lis = [10, 20, 30, 40, 40, 40, 70, 80, 99]

a = set(lis)
b = list(a)

maxx = 40
minn = 30

countt = 0
for i in b:
    if minn <= i <= maxx:
        countt += 1
        print(i)

print(countt)        
```

```python
def counelement(x):

    a = set(x)
    b = list(a)

    maxx = 40
    minn = 30

    countt = 0
    for i in b:
       if minn <= i <= maxx:
          countt += 1
          print(i)
    print(countt)        

lis = [10, 20, 30, 40, 40, 40, 70, 80, 99]
counelement(lis)
```
