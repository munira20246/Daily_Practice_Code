```python
# 24. Write a Python program to append a list to the second list.
a = [5, 6, 2, 3, 7, 10]
b = [36, 49, 64]
for i in a:
  b.append(i)
print(b)
```

```python
def join2list(x, y):
    for i in x:
        y.append(i)
    return y

b = [36, 49, 64]
a = [5, 6, 2, 3, 7, 10]
join2list(b, a)    
```
