```python
"""19.Write a Python program to get the difference between the two lists."""
a = [1, 3, 5, 7, 9, 11]
b = [5, 6, 2, 3, 7, 10]

z = list(set(a) - set(b))
y = list(set(b) - set(a))

d = z + y
print(d)
```

```python
def lisdifer(x, y):
    s = list(set(x) - set(y))
    t = list(set(y) - set(x))

    z = s + t
    return z
a = [1, 3, 5, 7, 9, 11]
b = [5, 6, 2, 3, 7, 10]

lisdifer(a, b)
```
