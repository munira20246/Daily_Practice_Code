```python
# 47. Write a Python program to insert an element before each element of a list.
lis0 = ['red', 'green', 'blue', 'yellow', 'white', 'purple']

# print(lis0)
for i in range(0, len(lis0)+(len(lis0)-1)):
    if i % 2 == 0:   
       lis0.insert(i, '*')
       
print(lis0)
```
