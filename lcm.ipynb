```python
def mul(x):
  lis = []
  for j in range(1, 100):
    j *= x
    lis.append(j)
  return lis

a = int(input("Enter 1st number---"))
b = int(input("Enter 2nd number---"))
c = int(input("Enter 2nd number---"))
lis1 = mul(a)
lis2 = mul(b)
lis3 = mul(c)
e = set(lis1)
f = set(lis2)
g = set(lis3)
uni_num = e & f & g
sort = sorted(uni_num)
lcm = sort[0]
print(f"The LCM of {a}, {b} and {c} is---{lcm}")
```

```python
def mul(x):
  lis = []
  for j in range(1, x+1):
    if x % j == 0:
        lis.append(j)
  return lis

a = int(input("Enter 1st number---"))
b = int(input("Enter 2nd number---"))
c = int(input("Enter 2nd number---"))
lis1 = mul(a)
lis2 = mul(b)
lis3 = mul(c)
e = set(lis1)
f = set(lis2)
g = set(lis3)
uni_num = e & f & g
g_sort = sorted(uni_num)
print(g_sort)
gcd = g_sort[-1]
print(f"The GCD of {a}, {b} and {c} is---{gcd}")
```
