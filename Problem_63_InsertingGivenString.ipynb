```python
# 63. Write a Python program to insert a given 
# string at the beginning of all items in a list.

num = [1, 2, 3, 4]
new_list = []
for i in num:
   n = "emp{0}".format(i)
   new_list.append(n)
print(new_list)   
    
```

```python
# Sample list : [1,2,3,4], string : emp
# Expected output : ['emp1', 'emp2', 'emp3', 'emp4']
```

```python
num = [1, 2, 3, 4]
new_list = ['emp{0}'.format(i) for i in num]
print(new_list)
```
