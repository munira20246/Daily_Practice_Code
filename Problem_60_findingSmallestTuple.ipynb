```python
# 60. Write a Python program to find a tuple, 
# the smallest second index value from a list of tuples.

x = [(4, 1), (1, 2), (6, 0)]
y = []
for i in x:
    y.append(i[1])
z = set(y)
a = list(z)
for j in x:
    if j[1] == a[0]:
        print(j)

```

```python
x = [(4, 1), (1, 2), (6, 0)]
print(min(x, key=lambda n: (n[1], n[0])))
```
