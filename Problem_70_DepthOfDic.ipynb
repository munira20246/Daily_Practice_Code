```python
# 70. Write a Python program to get the depth of a dictionary.
lis = ['abcd', 'abc', 'bcd', 'bkie', 'cder', 'cdsw', 'sdfsd', 'dagfa', 'acjd']
l = []
c = input("Enter a character---")
for item in lis:
    if item[0] == c:
        l.append(item)
print(f"Item start with {c} from the list:")
print(l)       
```
