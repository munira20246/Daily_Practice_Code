```python
""" 35. Write a Python program to create a list by concatenating a given list which
range goes from 1 to n."""

a = ['p', 'q']
n = 5
v = []
for i in range(1, n+1):
    v.append(a[0]+str(i))
    v.append(a[1]+str(i))
v    
```

```python
def concatenating(x):
    n = 5
    new = [f'{j}{i}' for i in range(1, n+1) for j in x]
    return new

a = ['p','q']
concatenating(a)
```
