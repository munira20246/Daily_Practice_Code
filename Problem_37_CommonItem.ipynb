```python
# 37. Write a Python program to find common items from two lists.

color1 = ["Red", "Green", "Orange", "White"]
color2 = ["Black", "Green", "White", "Pink"]
com = []

for i in color1:
    if i in color2:
        com.append(i)
print(com)        
```

```python
def findingCommon(a, b):
    comm = []
    for i in a:
        if i in b:
          comm.append(i)
    return comm      
c1 = ["Red", "Green", "Orange", "White"]
c2 = ["Black", "Green", "White", "Pink"]           

findingCommon(c1, c2)
```
