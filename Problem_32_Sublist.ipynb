```python
# 32. Write a Python program to check whether a list contains a sublist.

def is_sublist(main_list, sub_list):
    n = len(sub_list)
    if n == 0:
        return True 
    for i in range(len(main_list) - n + 1):
        if main_list[i:i+n] == sub_list:
            return True
    return False

main = [1, 2, 3, 4, 5, 6]
sub  = [3, 4, 5]
sub1  = [3, 4, 5, 10]
is_sublist(main, sub)
```
